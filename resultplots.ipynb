```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset variables (car, vote, nursery, breast cancer, abalone)
features = [6,16,8,9,8]
classes=[4,2,5,2,29]
```

```python
#optimization results

#RF, Matern kernel
RFM_cross = [0.984,0.970,0.993,0.770,0.261]
RFM_acc = [0.967,0.963,0.994,0.670,0.247]
RFM_i = [14,58,17,26,33]

#RF, custom kernel
RFC_cross = [0.993,0.970,0.993,0.760,0.262]
RFC_acc = [0.994, 0.963, 0.994, 0.674, 0.252]
RFC_i = [23,63,99,35,81]

#SVM, Matern kernel
SM_cross = [0.990,0.969,0.999,0.760,0.194]
SM_acc = [0.985,0.971,0.999,0.674,0.191]
SM_i = [18,65,18,88,47]

#SVM, custom kernel
SC_cross = [0.990,0.969,0.999,0.750,0.194]
SC_acc = [0.985,0.971,0.999,0.686,0.191]
SC_i = [24,37,15,78,21]

#KNN, Matern kernel
KM_cross = [0.908,0.914,0.950,0.720,0.225]
KM_acc = [0.927,0.957,0.950,0.663,0.239]
KM_i = [4,4,1,9,9]

#KNN, custom kernel
KC_cross = [0.908,0.914,0.950,0.720,0.225]
KC_acc = [0.927,0.957,0.950,0.663,0.239]
KC_i = [8,3,4,7,4]
```

```python
RFcdiff =[0,0,0,0,0]
RFadiff = [0,0,0,0,0]
RFidiff = [0,0,0,0,0]
Scdiff =[0,0,0,0,0]
Sadiff = [0,0,0,0,0]
Sidiff = [0,0,0,0,0]
Kidiff = [0,0,0,0,0]

for i in range(5):
  RFcdiff[i] = RFC_cross[i] - RFM_cross[i]
  RFadiff[i] = RFC_acc[i] - RFM_acc[i]
  RFidiff[i] = RFC_i[i] - RFM_i[i]
  Scdiff[i] = SC_cross[i] - SM_cross[i]
  Sadiff[i] = SC_acc[i] - SM_acc[i]
  Sidiff[i] = SC_i[i] - SM_i[i]
  Kidiff[i] = KC_i[i] - KM_i[i]

print(RFcdiff)
print(RFadiff)
print(RFidiff)
print(Scdiff)
print(Sadiff)
print(Sidiff)
print(Kidiff)
```
